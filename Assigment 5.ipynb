```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

alfa = 100
M=3
II = np.identity(M)
SS_0 = (1/alfa) * II
Ntotal = N1 + N2

def probablityoftstart1(amap,E):
    
    P = sigmoid( np.asscalar(amap)/(1 + ((np.pi/8)* np.asscalar(E))**(1/2)))
    return P

def sigmoid(x):
    sig = 1 / (1 + (np.exp(-x)))
               
    return sig

def sigmoid_bar(w_old, X):
    sigmoidbar = np.matrix(np.zeros([Ntotal,1]))
    for i in range(Ntotal):
        a= np.matmul(w_old, X[i].transpose())
      
        sigmoidbar[i,0]=sigmoid(a)
      
    return sigmoidbar
       

def amap(w, x):
  
    amap = np.matmul(w, x)
    
    return amap

def sigmaE(Xf, SS_0, w_old):
    
 
    t =np.matrix(T[:,1]).transpose()
    
  
    E1=np.matmul(SS_0, w_old.transpose())
    s= sigmoid_bar(w_old, Xf)
    E3 = s-t
    E2=np.matmul(Xf.transpose(), E3)
    E= E1 +E2

    return E

def sigmaEn(x, SS):
    
    E1=np.matmul(x.T, SS)
    E= np.matmul(E1, x)
   
    return E
    
def calSS (X, R):
    SS_tmp = np.matmul(X.transpose(), R)
    SS = np.matmul(SS_tmp, X)
    
    return SS

def calH(SS_0, X, R):
    H_tmp = np.matmul(X.transpose(), R)
    H_tmp = np.matmul(H_tmp, X)
    H = SS_0 + H_tmp
    
    return H

def calR(X, w_old):
    R = np.matrix(np.zeros([Ntotal, Ntotal]))
    for i in range(Ntotal):
    
        inner = np.matmul(w_old, X[i].transpose())
        R[i, i] = (sigmoid(inner) * (1- sigmoid(inner)))   
    return R

def diff(w_old, w_new):
    sum =0
    for i in range(M):
        a = np.absolute(w_old[0,i] - w_new[0,i])
        sum = sum + a
    return sum    
        
    
        
def calw(Xf):
    
    w_old = np.matrix(np.zeros([1, 3]))
    R = calR(Xf, w_old)
    E=sigmaE(Xf, SS_0, w_old)
    
    H=calH(SS_0, Xf, R)
    epsilon = 0.0000001
    diffrence= 10
    while abs(diffrence)>epsilon:
        
     
        z= np.matmul(np.linalg.inv(H), E)
        b = z.transpose()
        w_new = w_old - b
        diffrence=diff(w_old, w_new)
        w_old= w_new
        R = calR(Xf, w_old)
        E=sigmaE(Xf, SS_0, w_old)
        H=calH(SS_0, Xf, R)

    return w_new   
        
        
def predictive(i, j, w, SS):
 
    x = np.array([1, i, j])
    a = amap(w, x)
    En = sigmaEn(x, SS)
    P = probablityoftstart1(a,En)
    return P


Arrayof1 = np.matrix(np.ones(Ntotal))
AT= Arrayof1.transpose()
Xf = np.concatenate ((AT, X),axis=1)
w = calw(Xf)


R = calR(Xf, w)
SS = calSS(Xf, R)
SS =np.linalg.inv(SS)
print(SS)
BIGMATRIX = np.matrix(np.empty([30,30]))
for i in range(30):
    for j in range(30):
        b = predictive(i/3,j/3,w,SS)
        BIGMATRIX[i, j] = b


plt.scatter(X[:, 0], X[:, 1], marker ='o', c=color)
plt.imshow(BIGMATRIX.T, interpolation='bilinear', origin='lower', cmap=plt.cm.jet, extent = (0,10,0,10))
plt.colorbar()
plt.show()
```
